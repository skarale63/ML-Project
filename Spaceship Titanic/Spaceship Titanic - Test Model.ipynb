```python
import pandas as pd
import numpy as np
import pickle
```

## Data Preparation for testing

```python
df = pd.read_csv('test.csv')
df
```

```python
df.info()
```

```python
# Filling NA values by mode of variable
df['HomePlanet'].fillna(value=df['HomePlanet'].mode()[0],inplace=True)
df['CryoSleep'].fillna(value=df['CryoSleep'].mode()[0],inplace=True)
df['Cabin'].fillna(value=df['Cabin'].mode()[0],inplace=True)
df['Destination'].fillna(value=df['Destination'].mode()[0],inplace=True)
df['VIP'].fillna(value=df['VIP'].mode()[0],inplace=True)
df['Name'].fillna(value=df['Name'].mode()[0],inplace=True)

# Filling NA values by mean of variable
df['Age'].fillna(value=df['Age'].mean(),inplace=True)
df['RoomService'].fillna(value=df['RoomService'].mean(),inplace=True)
df['FoodCourt'].fillna(value=df['FoodCourt'].mean(),inplace=True)
df['ShoppingMall'].fillna(value=df['ShoppingMall'].mean(),inplace=True)
df['Spa'].fillna(value=df['Spa'].mean(),inplace=True)
df['VRDeck'].fillna(value=df['VRDeck'].mean(),inplace=True)
```

```python
df.info()
```

```python
df_new = df.drop(['PassengerId','Name','Cabin'],axis=1)
```

```python
df_new
```

```python
# Applying One Hot Encoding for Categorical features...
df1=pd.get_dummies(df_new['HomePlanet'],prefix='HomePlanet',drop_first=True)
df2=pd.get_dummies(df_new['Destination'],prefix='Destination',drop_first=True)
df_new =pd.concat([df_new,df1,df2],axis=1)
df_new = df_new.drop(['HomePlanet','Destination'],axis=1)

# Replace True with 1 and False with 0 
df_new['CryoSleep'] = np.where(df_new['CryoSleep']==True,1,0)
df_new['VIP'] = np.where(df_new['VIP']==True,1,0)
```

```python
df_new
```

## Data testing on build model

```python
filename = 'xgboost_model.pickle'
filename_scaler = 'scaler_model.pickle'
scaler_model = pickle.load(open(filename_scaler, 'rb'))
Xgb_model = pickle.load(open(filename, 'rb'))
```

#### Prediction on given test.csv data

```python
x_test=scaler_model.transform(df_new)
y_pred=Xgb_model.predict(x_test)
```

#### Now creating submission file

```python
df_sub = pd.DataFrame()
df_sub['PassengerId'] = df['PassengerId']
df_sub['Transported'] = np.where(y_pred==1,True,False)
df_sub.set_index("PassengerId", inplace = True)
```

#### Saving file in csv format

```python
df_sub.to_csv('submission.csv')
```
